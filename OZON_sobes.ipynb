### Q2: 
А сколько в среднем проходит между тем как пользователь получил первый заказ, и сделал следующий заказ?

Для решения этой задачи предлагаю определить самую раннюю и последнюю дату, то есть для каждого пользователя определю MIN и MAX даты и узнаем сколько каждый пользователь оофрмлял заказов за этот период, но перед нахождением среднего надо вычесть из разности ранней и поздней дат время ожидания заказов, но мы уже имеем несколько заготовок в виде таблиц **df_time_to_delivery** отсортированной по каждому пользователю

```python
## то есть смысл нахождения AVG() такой, допустим пользователь оформил два заказа
## первый заказ в 12:10 и получил его в 12:40
## второй заказ в 15:45 и получил его в 16:00

## Мы прекрасно понимаем что по условию задачи нам нужно найти время (15:45 - 12:40) и получить ответ: 3 часа 5 минут

## но в sql это сделать можно следующим образом: находим разницу между ранней и поздней датами => (16:00 - 12:10) = 3 часа 50 минут

## теперь из этой разницы вычтем сумму всех ожиданий клиента => 3часа 50 минут  - ((12:40 - 12:10 = 30 минут) + 
## + (16:00 - 15:45 = 15 минут) = 45 минут) = 3 часа 5 минут

## Значит задача решена верно, применим это уже с реальными данными
```

```python
df_user_id = pd.read_sql_query("SELECT user_id, ROUND((julianday(delivered_at) - julianday(created_at))*24, 2) AS time_to_delivery_user_id FROM orders ORDER BY user_id DESC LIMIT 10", con)
df_user_id
```

```python
df_user_id = pd.read_sql_query("SELECT user_id, MIN(created_at), MAX(delivered_at), ROUND((julianday(MAX(delivered_at))-julianday(MIN(created_at)))*24,2) AS diff_max_man,COUNT(user_id) AS cnt, sum(ROUND((julianday(delivered_at) - julianday(created_at))*24, 2)) AS sum_time_to_delivery_user_id FROM orders GROUP BY user_id ORDER BY user_id DESC LIMIT 10", con)
df_user_id
```

Данные получены, теперь нужно лишь найти разницу столбцов и поделить на колличество заказов

Можно заметить, что пользователей с одним заказом нужно удалить из данных, так как по сути у него нет той метрики, которую просят найти в задаче, но их данные в столбцах **diff_max_man** и **sum_time_to_delivery_user_id** равны и в итоге они получат 0 в ячейке, что будет означать отсутствие других закозов после первого

```python
df_user_id = pd.read_sql_query("SELECT user_id,((diff_max_man-sum_time_to_delivery_user_id)/cnt) AS avg_time FROM(SELECT user_id, MIN(created_at), MAX(delivered_at), ROUND((julianday(MAX(delivered_at))-julianday(MIN(created_at)))*24,2) AS diff_max_man,COUNT(user_id) AS cnt, sum(ROUND((julianday(delivered_at) - julianday(created_at))*24, 2)) AS sum_time_to_delivery_user_id FROM orders GROUP BY user_id ORDER BY user_id DESC) LIMIT 10", con)
df_user_id
```

Для наглядности взглянем на эту таблицу без нулей

```python
df_user_id = pd.read_sql_query("SELECT user_id,((diff_max_man-sum_time_to_delivery_user_id)/cnt) AS avg_time FROM(SELECT user_id, MIN(created_at), MAX(delivered_at), ROUND((julianday(MAX(delivered_at))-julianday(MIN(created_at)))*24,2) AS diff_max_man,COUNT(user_id) AS cnt, sum(ROUND((julianday(delivered_at) - julianday(created_at))*24, 2)) AS sum_time_to_delivery_user_id FROM orders GROUP BY user_id ORDER BY user_id DESC) WHERE diff_max_man != sum_time_to_delivery_user_id LIMIT 10", con)
df_user_id
```
